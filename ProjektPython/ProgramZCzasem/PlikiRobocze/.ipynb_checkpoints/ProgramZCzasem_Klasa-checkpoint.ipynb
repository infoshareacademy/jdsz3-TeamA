```python
class ProgramZCzasem:
    
    kolumnyInt = ["Clothing ID","Age","Rating","Recommended IND","Positive Feedback Count"]
    parametrDodatkowy = ''
    
    menu = {
        '' :  {'typ' : 'menu',
               'oczekiwanaOdp' : 'menu_int',
               'tekst' : 
                   '\nMenu początkowe: \
                    \n1 - Ogólne informacje o danych \
                    \n2 - Szczegółowe informacje o kolumnach typu integer \
                    \n3 - Szczegółowe informacje o kolumnach typu string \
                    \nW każdym momencie wybierz: \
                    \nZ - aby przejść do poprzedniego menu \
                    \nX - aby wyjść z programu'},
        '1' : {'typ' : 'menu',
               'oczekiwanaOdp' : 'menu_int',
               'tekst' : 
                   '\nWybrałeś ogólne informacje o danych. \
                    \nWybierz interesujące Cię informacje: \
                    \n1 - Opis danych \
                    \n2 - Pierwsze wiersze tabeli \
                    \n3 - Ogólne statystyki opisowe \
                    \n4 - Podsumowanie danych'},
        '11' : {'typ' : 'akcja',
               'akcja' : 'akcja_11'},
        '12' : {'typ' : 'akcja',
               'akcja' : 'akcja_12'},
        '13' : {'typ' : 'akcja',
               'akcja' : 'akcja_13'},
        '14' : {'typ' : 'akcja',
               'akcja' : 'akcja_14'},
        '2' : {'typ' : 'menu',
               'oczekiwanaOdp' : 'zmienna_int',
               'tekst' : 
                   '\nŚwietnie! \
                   \nCzyli chcesz wiedzieć więcej o liczbach w naszej tabeli, a oto i dostępne kolumny: \
                   \n1 - Clothing ID \
                   \n2 - Age \
                   \n3 - Rating \
                   \n4 - Recommended IND \
                   \n5 - Positive Feedback Count'},
        '20' : {'typ' : 'menu',
                'oczekiwanaOdp' : 'menu_int',
                'tekst' :
                    f'\nWybrana kolumna to {parametrDodatkowy} \
                     \nJakiej informacji potrzebujesz ? \
                     \n1 - Wszystkie dane z kolumny \
                     \n2 - Szczegółowe dane statystyczne \
                     \n3 - Wykres skrzynkowy \
                     \n4 - Rozkład obserwacji \
                     \n5 - Jądrowy estymator gęstości'},
        '201' : {'typ' : 'akcja',
               'akcja' : 'akcja_201'},
        '202' : {'typ' : 'akcja',
               'akcja' : 'akcja_202'},
        '203' : {'typ' : 'akcja',
               'akcja' : 'akcja_203'},
        '204' : {'typ' : 'akcja',
               'akcja' : 'akcja_204'},
        '205' : {'typ' : 'akcja',
               'akcja' : 'akcja_205'},
        '3' : {'typ' : 'menu',
                'oczekiwanaOdp' : 'menu_int',
                'tekst' : 
                   '\nChcesz się bliżej przyjrzeć komentarzom? Super! \
                    \nPrzed Tobą kilka możliwości: \
                    \n1 - Informacje statystyczne o komentarzach \
                    \n2 - Przeczytaj konkretny komentarz \
                    \n3 - Wyszukaj komentarze zawierające konkretne słowo \
                    \n4 - Zobacz chmurę najpopularniejszych słów w komentarzach \
                    \n5 - Przetestuj biblioteki oceniające, czy dany tekst jest pozytywny, czy negatywny, wpisując własną sentencje'},
        '31' : {'typ' : 'akcja',
                'akcja' : 'akcja_31'},
        '32' : {'typ' : 'menu',
                'oczekiwanaOdp' : 'zmienna_int',
                'tekst' : 'Podaj numer z przedziału (1-23485), aby przeczytać konkretny komentarz.'},
        '320' : {'typ' : 'akcja',
                'akcja' : 'akcja_320'},

        '33' : {'typ' : 'menu',
                'oczekiwanaOdp' : 'zmienna_string',
                'tekst' : 'Podaj słowo kluczowe.'},
        '330' : {'typ' : 'akcja',
                'akcja' : 'akcja_330'},
        '34' : {'typ' : 'akcja',
                'akcja' : 'akcja_34'},
        '35' : {'typ' : 'menu',
                'oczekiwanaOdp' : 'zmienna_string',
                'tekst' : 'Podaj tekst.'},
        '350' : {'typ' : 'akcja',
                'akcja' : 'akcja_350'},
        'X' : {'typ' : 'menu',
               'tekst' : 'Wychodzę z programu. Do widzenia.',
               'oczekiwanaOdp' : ''
              }
    }
    
    
    def __init__(self, sciezkaDoDanych):
        self.dane = pd.read_csv(sciezkaDoDanych)
        self.ProgramDziala = True
        self.pozycjaMenu = []
        self.oczekiwanaOdp = 'menu_int'
    
    def poprosOOdp(self):
        self.odpowiedz = input('\n################Twoj wybor:')
        
    def zwalidujOdpowiedz(self):
        if self.odpowiedz.upper() == 'X':
            self.pozycjaMenu = ['X']
            self.ProgramDziala = False
        elif self.odpowiedz.upper() == 'Z':
            try:
                del self.pozycjaMenu[-1]
                print('\nIdę poziom wyżej.')
            except IndexError:
                print('\nNie możesz iść wyżej. Wyświetlę dla Ciebie menu początkowe jeszcze raz.')
        elif self.oczekiwanaOdp == 'menu_int':
            try:
                self.odpowiedz = int(self.odpowiedz)
                self.pozycjaMenu.append(self.odpowiedz)
                print(f'\nTwój wybór to: {self.odpowiedz}')
            except:
                print('\nUps! Podałeś niepoprawną odpowiedź. Wyświetlę dla Ciebie menu jeszcze raz.')
        elif self.oczekiwanaOdp == 'zmienna_int':
            try:
                self.odpowiedz = int(self.odpowiedz)
                self.parametrDodatkowy = self.odpowiedz
                self.pozycjaMenu.append(0)
                print(f'\nTwój wybór to: {self.odpowiedz}')
            except:
                print('\nUps! Podałeś niepoprawną odpowiedź. Wyświetlę dla Ciebie menu jeszcze raz.')
        elif self.oczekiwanaOdp == 'zmienna_string':
            try:
                self.parametrDodatkowy = self.odpowiedz
                self.pozycjaMenu.append(0)
                print(f'\nTwój wybór to: {self.odpowiedz}')
            except:
                print('\nUps! Podałeś niepoprawną odpowiedź. Wyświetlę dla Ciebie menu jeszcze raz.')
        
        pozycjaMenuString = ''.join([str(i) for i in self.pozycjaMenu])
        if not pozycjaMenuString in self.menu.keys():
            del self.pozycjaMenu[-1]
            print('\nUps! Podałeś niepoprawną odpowiedź. Wyświetlę dla Ciebie menu jeszcze raz.')

    def Wykonaj(self):
        #sprawdz czy moze to byc atrybut
        pozycjaMenuString = ''.join([str(i) for i in self.pozycjaMenu])
        if self.menu[pozycjaMenuString]['typ'] == 'menu':
            print(f'Poziom menu: {len(pozycjaMenuString) + 1}')
            print(self.menu[pozycjaMenuString]['tekst'])
            self.oczekiwanaOdp = self.menu[pozycjaMenuString]['oczekiwanaOdp']
        elif self.menu[pozycjaMenuString]['typ'] == 'akcja':
            self.WykonajAkcje(self.menu[pozycjaMenuString]['akcja'])
            del self.pozycjaMenu[-1]
            self.Wykonaj()
    
    def WykonajAkcje(self, akcja):
        if akcja == 'akcja_11':
            print("Wybrałeś opis danych:\n")
            print("Dane pochodzą ze strony: \n'www.kaggle.com/nicapotato/womens-ecommerce-clothing-reviews'\n")        
            print("Jest to zestaw danych E-Commerce dotyczący odzieży damskiej, zawierający recenzje klientów.")
            print("Z racji tego, że są to prawdziwe dane handlowe, zostały zanonimizowane, a odniesienia do firmy w tekście i treści przeglądu zostały zastąpione nazwą'sprzedawca detaliczny'\n")
            print("Ten zestaw danych zawiera 23486 wierszy i 10 zmiennych cech. Każdy wiersz odpowiada opinii klienta i zawiera zmienne:")
            print("Identyfikator odzieży, wiek klienta, tytuł komentarza, tekst komentarza, ocena, informacja o poleceniu/niepoleceniu produktu, liczba pozytywnych opinii, nazwa grupy, działu, kategorii ubrania\n")
        elif akcja == 'akcja_12':
            print("Wybrałeś podgląd pierwszych wierszy tabeli:\n")
            print(self.dane.head())
        elif akcja == 'akcja_13':
            print("Wybrałeś ogólne statystyki opisowe:\n")
            print(self.dane.describe())
        elif akcja == 'akcja_14':
            print("Wybrałeś podsumowanie danych:\n")
            print(self.dane.info())
        elif akcja == 'akcja_201':
            try:
                print("LP.",self.kolumnyInt[self.parametrDodatkowy-1])
                print(self.dane[self.kolumnyInt[self.parametrDodatkowy-1]])
            except IndexError:
                print('Wybrałeś niepoprawną kolumnę. Wróć do poprzedniego menu i wybierz poprawny numer kolumny.')
        elif akcja == 'akcja_202':
            try:    
                print(self.dane.loc[:,self.kolumnyInt[self.parametrDodatkowy-1]].describe())
            except IndexError:
                print('Wybrałeś niepoprawną kolumnę. Wróć do poprzedniego menu i wybierz poprawny numer kolumny.')
        elif akcja == 'akcja_203':
            try:
                color = dict(markerfacecolor='r', marker='s')
                fig2, ax2 = plt.subplots()
                ax2.set_title(self.kolumnyInt[self.parametrDodatkowy-1])
                ax2.boxplot(self.dane[self.kolumnyInt[self.parametrDodatkowy-1]], notch=True, vert=False, flierprops=color)
                plt.show()
            except IndexError:
                print('Wybrałeś niepoprawną kolumnę. Wróć do poprzedniego menu i wybierz poprawny numer kolumny.')
        elif akcja == 'akcja_204':
            try:
                sns.distplot(self.dane[self.kolumnyInt[self.parametrDodatkowy-1]])
                plt.show()
            except IndexError:
                print('Wybrałeś niepoprawną kolumnę. Wróć do poprzedniego menu i wybierz poprawny numer kolumny.')
        elif akcja == 'akcja_205':
            try:
                sns.distplot(self.dane[self.kolumnyInt[self.parametrDodatkowy-1]], kde=True,kde_kws={"color": "b", "alpha":0.3, "linewidth": 5, "shade":True })
                plt.show()
            except IndexError:
                print('Wybrałeś niepoprawną kolumnę. Wróć do poprzedniego menu i wybierz poprawny numer kolumny.')
        elif akcja == 'akcja_31':
            print('Jeszcze nad tym pracujemy!')
        elif akcja == 'akcja_320':
            print('Jeszcze nad tym pracujemy!')
        elif akcja == 'akcja_330':
            print('Jeszcze nad tym pracujemy!')
        elif akcja == 'akcja_34':
            print('Jeszcze nad tym pracujemy!')
        elif akcja == 'akcja_350':
            print('Jeszcze nad tym pracujemy!')
        else:
            print('Coś poszło nie tak jak powinno. Wyświetlę dla Ciebie menu jeszcze raz.')
```
