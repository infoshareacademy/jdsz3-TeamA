```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

# Sales of Child Car Seats

     - a simulated data set containing sales of child car seats at 400 different stores

```python
filepath = r'https://raw.githubusercontent.com/infoshareacademy/jdsz3-zczasem/ProjektML/ProjektML/Dane/Carseats_Clean_Final.csv'
```

- Sales - Unit sales (in thousands) at each location
    - CompPrice - Price charged by competitor at each location
    - Income - Community income level (in thousands of dollars)
    - Advertising - Local advertising budget for company at each location (in thousands of dollars)
    - Population - Population size in region (in thousands)
    - Price - Price company charges for car seats at each site
    - ShelveLoc - A factor with levels Bad, Good and Medium indicating the quality of the shelving location for   
      the car seats at each site
    - Age - Average age of the local population
    - Education - Education level at each location
    - Urban - A factor with levels No and Yes to indicate whether the store is in an urban or rural location
    - US - A factor with levels No and Yes to indicate whether the store is in the US or not

```python
df = pd.read_csv(filepath)
df.head()
```

### Multiple regression model - all variables

```python
dfAll = df.drop(['Sales_class'], axis = 1)
```

```python
trainSetAll, testSetAll = train_test_split(dfAll, test_size = 0.2, random_state = 9)
```

```python
mlRegAll = ols('Sales ~ CompPrice + Income + Advertising + Population + Price + Age + ShelveLoc + Education + Urban + US', trainSetAll).fit()
print(mlRegAll.summary())
```

##### Score of the multiple regression model - all variables

```python
mlRegAll_r2score = r2_score(testSetAll.Sales, mlRegAll.predict(testSetAll))
print(mlRegAll_r2score)
```

### Final multiple regression model 
Removed variables: Population, Education, Urban, US, ShelveLoc

Added variables: ShelveLoc^2

```python
dfFinal = df.drop(['Population', 'Education', 'Urban', 'US', 'Sales_class'], axis = 1)
dfFinal['ShelveLoc2'] = dfFinal['ShelveLoc'] ** 2
dfFinal = dfFinal.drop(['ShelveLoc'], axis = 1)
```

```python
trainSetFinal, testSetFinal = train_test_split(dfFinal, test_size = 0.2, random_state = 9)
```

```python
mlRegFinal = ols('Sales ~ CompPrice + Income + Advertising + Price + Age + ShelveLoc2', trainSetFinal).fit()
print(mlRegFinal.summary())
```

##### Score of the final multiple regression model

```python
mlRegFinal_r2score = r2_score(testSetFinal.Sales, mlRegFinal.predict(testSetFinal))
print(mlRegFinal_r2score)
```
