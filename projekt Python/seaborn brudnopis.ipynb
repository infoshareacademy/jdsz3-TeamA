```python

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Womens Clothing E-Commerce Reviews.csv") # wczytanie pliku z danymi
df['Class Name NUM'] = df['Class Name'].astype('category').cat.codes
df.head() # wypisanie części pliki w postaci tabeli
```

## Boxploty na wiek i ocenę

```python
color = dict(markerfacecolor='r', marker='s')
fig2, ax2 = plt.subplots()
ax2.set_title('Rozkład wieku respondentów')
ax2.boxplot(df["Age"], notch=True, vert=False, flierprops=color)
plt.show()
```

```python
ax2.set_title('Rozkład wieku ocen')
ax2.boxplot(df["Rating"], notch=True, vert=False, flierprops=color)
fig2, ax3 = plt.subplots()
ax3.boxplot(df["Rating"], notch=True, vert=False, flierprops=color)
plt.show()
```

```python

df['Division Name']=df['Division Name'].astype('category').cat.codes
df['Department Name']=df['Department Name'].astype('category').cat.codes
df.corr()
sns.pairplot(df[['Age','Rating','Class Name NUM','Recommended IND','Department Name','Division Name']], kind="reg")
plt.show()
```

```python
df['Class Name']=df['Class Name'].astype('category').cat.codes
plt.scatter(df[['Age']], df[['Class Name']], s=df[['Rating']]*25, c="red", cmap="Blues", alpha=0.4, edgecolors="grey", linewidth=2)
 
# Add titles (main and on axis)
plt.xlabel("Age")
plt.ylabel("Class Name")
plt.title("Korelacja")
 
plt.show()
```
